````python
import pandas as pd

# Criação do DataFrame
data = {
    'name': ['Afghanistan', 'Albania', 'Algeria', 'Andorra', 'Angola'],
    'continent': ['Asia', 'Europe', 'Africa', 'Europe', 'Africa'],
    'area': [652230, 28748, 2381741, 468, 1246700],
    'population': [25500100, 2831741, 37100000, 78115, 20609294],
    'gdp': [20343000000, 12960000000, 188681000000, 3712000000, 100990000000]
}

df = pd.DataFrame(data)
print(df)
````

A country is big if:
it has an area of at least three million (i.e., 3000000 km2), or
it has a population of at least twenty-five million (i.e., 25000000).

>> Write a solution to find the name, population, and area of the big countries.

Return the result table in any order.

````python
df.head()
````

````python
# selecionando apenas as colunas solicitadas no enunciado

colunas_desejadas = ['name', 'population', 'area']
df_filtrado = df[colunas_desejadas]
df_filtrado.head()
````

````python
# aplicando os filtros solicitados no enunciado

# condicoes
# cria uma série booleana (True/False) para cada linha do DataFrame, indicando se a condição de área é satisfeita
area_filter         = df_filtrado['area'] >= 3000000
population_filter   = df_filtrado['population'] >= 25000000

# Filtrando o DataFrame com OR
df_is_big = df_filtrado[area_filter | population_filter]
df_is_big.head()
````

````python
# validando/checando os filtros
print(df_filtrado[population_filter])
````

````python
# validando/checando os filtros
print(df_filtrado[area_filter])
````

````python
# SOLUCAO FINAL

# 1️⃣ Defina a condição para países grandes
cond_area = df['area'] >= 3000000
cond_pop = df['population'] >= 25000000

# 2️⃣ Filtre o DataFrame usando a condição OR
df_filtrado = df[cond_area | cond_pop]

# 3️⃣ Selecione apenas as colunas desejadas
df_result = df_filtrado[['name', 'population', 'area']]

# 4️⃣ Retorne o DataFrame final
df_result.head()
````
