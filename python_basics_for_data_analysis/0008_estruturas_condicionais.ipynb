Dada a lista de produtos abaixo, imprima a faixa de pre√ßo de cada um com base no valor:

* At√© 10 ‚Üí **Barato**
* Entre 10 e 100 ‚Üí **M√©dio**
* Acima de 100 ‚Üí **Caro**

```python
produtos = [
    {'nome': 'Caneta', 'preco': 2.5},
    {'nome': 'Headphone', 'preco': 150},
    {'nome': 'Mouse', 'preco': 75},
    {'nome': 'Monitor', 'preco': 200}

```

üëâ Escreva o c√≥digo que percorre a lista e imprime o nome do produto e sua classifica√ß√£o de pre√ßo.

Me avise quando tiver pronto ou se quiser que eu valide sua resposta.


````python
# Dada a lista de produtos abaixo, imprima a faixa de pre√ßo de cada um com base no valor:
# * At√© 10 ‚Üí **Barato**
# * Entre 10 e 100 ‚Üí **M√©dio**
# * Acima de 100 ‚Üí **Caro**

produtos = [
    {'nome': 'Caneta', 'preco': 2.5},
    {'nome': 'Headphone', 'preco': 150},
    {'nome': 'Mouse', 'preco': 75},
    {'nome': 'Monitor', 'preco': 200}
]

for produto in produtos:
    preco = produto['preco']
    nome = produto['nome']

    if preco > 100:
        faixa = 'Caro'
    elif preco >= 10:
        faixa = 'M√©dio'
    else:
        faixa = 'Barato'

    print(f"Produto: {nome}, Pre√ßo: {preco:.2f}‚Ç¨, Faixa: {faixa}")

````

## üß† Como pensar em **if dentro de for** ‚Äî passo a passo para racioc√≠nio l√≥gico

---

### 1. Entenda o problema:

Voc√™ tem uma **lista** de produtos e quer **avaliar cada produto individualmente** para classificar o pre√ßo.

---

### 2. Pense no que √© repetido:

Para cada produto da lista, voc√™ precisa:

* Pegar o pre√ßo;
* Aplicar regras (condi√ß√µes) para decidir a faixa (barato, m√©dio, caro);
* Imprimir o resultado.

---

### 3. O que √© iterar com o `for`:

Um loop `for` serve para **percorre cada item da lista** e executar algo para cada item.

```python
for produto in produtos:
    # aqui voc√™ est√° trabalhando com 1 produto por vez
```

---

### 4. Dentro do `for`, aplique as regras:

Cada produto tem um pre√ßo, e para esse pre√ßo, voc√™ usa `if`, `elif`, `else` para definir a faixa.

```python
if preco > 100:
    faixa = 'Caro'
elif preco >= 10:
    faixa = 'M√©dio'
else:
    faixa = 'Barato'
```

---

### 5. Depois de decidir a faixa, fa√ßa a a√ß√£o final, que no caso √© imprimir.

---

### **Resumo do fluxo mental:**

1. Para cada produto na lista ‚Üí **`for`**
2. Pega o pre√ßo ‚Üí `preco = produto['preco']`
3. Decide faixa com condi√ß√µes ‚Üí `if ... elif ... else`
4. Imprime resultado formatado ‚Üí `print(f'...')`

---

### üí° Dica para treinar:

* Sempre pergunte: "Eu preciso repetir uma a√ß√£o para v√°rios itens?" ‚Üí Use `for`.
* Dentro da repeti√ß√£o, se quiser tomar decis√µes diferentes ‚Üí Use `if`.
* Separe a l√≥gica em passos simples, por exemplo, calcule um valor, depois decida algo, depois imprima.

### üìä **Exerc√≠cio 6: M√©dia da turma e status individual**

Dada a lista de alunos com suas respectivas notas, calcule a **m√©dia geral da turma** e, para cada aluno:

* Informe se ele est√° **acima**, **na m√©dia**, ou **abaixo da m√©dia geral**

```python
alunos = [
    {'nome': 'Ana', 'nota': 8.5},
    {'nome': 'Bruno', 'nota': 6.2},
    {'nome': 'Carlos', 'nota': 7.8},
    {'nome': 'Diana', 'nota': 9.1},
    {'nome': 'Eduardo', 'nota': 5.4}
]
```

üì• **Sa√≠da esperada (exemplo):**

```
M√©dia da turma: 7.4

Aluno: Ana - Nota: 8.5 - Acima da m√©dia
Aluno: Bruno - Nota: 6.2 - Abaixo da m√©dia
...
```

Quando estiver pronto, envie sua solu√ß√£o.


````python
alunos = [
    {'nome': 'Ana', 'nota': 8.5},
    {'nome': 'Bruno', 'nota': 6.2},
    {'nome': 'Carlos', 'nota': 7.8},
    {'nome': 'Diana', 'nota': 9.1},
    {'nome': 'Eduardo', 'nota': 5.4}
]

soma = 0

for aluno in alunos:
    soma += aluno['nota']

media = soma / len(alunos)

print(f'N√∫mero de alunos: {len(alunos)}, Soma das notas: {soma:.2f}')
print(f'M√©dia da turma: {media:.2f}')

print() # separador

for aluno in alunos:
    nome = aluno['nome']
    nota = aluno['nota']

    if nota >= media:
        situacao = 'Acima da m√©dia'
    else:
        situacao = 'Abaixo da m√©dia'

    print(f'Aluno: {nome} - Nota: {nota:.2f} - Situa√ß√£o: {situacao}')

````
